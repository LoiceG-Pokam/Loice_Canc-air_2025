```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from libpysal import weights
from esda import Moran, Moran_Local
import seaborn as sns
from matplotlib.patches import Rectangle

# ===== 1. CHARGEMENT DES DONNÉES =====
# Charger le CSV avec les clusters
df_clusters = pd.read_csv("R:/Direction_Data/0_Projets/Projet_CANCAIR/2025_Projet_Loice/Data/df_patients_pca_clusters.csv", sep=";")

# Afficher les premières lignes et les colonnes pour vérification
print("Aperçu des données CSV:")
print(df_clusters.head())
print(f"\nColonnes disponibles: {list(df_clusters.columns)}")

# ADAPTATION NÉCESSAIRE: Ajustez les noms des colonnes selon votre CSV
# Exemples de noms possibles à remplacer:
# - 'longitude', 'lon', 'x', 'X'  
# - 'latitude', 'lat', 'y', 'Y'
# - 'cluster', 'cluster_id', 'groupe', 'classe'

# ⚠️ MODIFIEZ CES NOMS DE COLONNES SELON VOTRE CSV ⚠️
col_longitude = 'x'  # Remplacez par le nom de votre colonne longitude
col_latitude = 'y'   # Remplacez par le nom de votre colonne latitude  
col_cluster = 'cluster'     # Remplacez par le nom de votre colonne cluster

# Vérifier que les colonnes existent
required_cols = [col_longitude, col_latitude, col_cluster]
missing_cols = [col for col in required_cols if col not in df_clusters.columns]
if missing_cols:
    print(f"ERREUR: Colonnes manquantes: {missing_cols}")
    print("Veuillez ajuster les noms des colonnes dans le script.")
    exit()

# Supprimer les lignes avec des valeurs manquantes pour les coordonnées
df_clusters = df_clusters.dropna(subset=[col_longitude, col_latitude, col_cluster])

print(f"Nombre d'observations après nettoyage: {len(df_clusters)}")

# Créer le GeoDataFrame à partir du CSV
gdf_clusters = gpd.GeoDataFrame(
    df_clusters,
    geometry=gpd.points_from_xy(df_clusters[col_longitude], df_clusters[col_latitude]),
    crs="EPSG:4326"  # WGS84 (coordonnées géographiques classiques)
)

# Reprojeter en Lambert 93 (système projeté pour la France)
gdf_clusters = gdf_clusters.to_crs("EPSG:2154")

# Charger les contours administratifs d'Île-de-France
gdf_idf_region = gpd.read_file(r"H:\PFE Loice\Notebooks\data\shapefile\IDF.shp")
gdf_idf_departements = gpd.read_file(r"R:\Direction_Data\0_Projets\Projet_CANCAIR\data\zones_geographiques\departements\IDF\dpt_idf.shp")

# Vérifier les systèmes de coordonnées et reprojeter si nécessaire
if gdf_idf_region.crs != 'EPSG:2154':
    gdf_idf_region = gdf_idf_region.to_crs('EPSG:2154')
if gdf_idf_departements.crs != 'EPSG:2154':
    gdf_idf_departements = gdf_idf_departements.to_crs('EPSG:2154')

# Filtrer les points dans la région IDF si nécessaire
# (optionnel - si votre CSV contient des points hors IDF)
gdf_clusters_idf = gpd.overlay(gdf_clusters, gdf_idf_region, how='intersection', keep_geom_type=False)
if len(gdf_clusters_idf) < len(gdf_clusters):
    print(f"Points filtrés en IDF: {len(gdf_clusters_idf)}/{len(gdf_clusters)}")
    gdf_clusters = gdf_clusters_idf

print(f"Nombre d'observations: {len(gdf_clusters)}")
print(f"Clusters présents: {sorted(gdf_clusters[col_cluster].unique())}")
print(f"Étendue spatiale: {gdf_clusters.total_bounds}")

# ===== 2. CRÉATION DE LA MATRICE DE POIDS SPATIALE =====
# Distance de voisinage (ajustez selon votre contexte)
distance_threshold = 500  # 5 km en mètres

# Création de la matrice de poids basée sur une distance fixe
w = weights.DistanceBand.from_dataframe(gdf_clusters, threshold=distance_threshold, binary=True)

# Normalisation en ligne (chaque ligne somme à 1)
w.transform = 'R'

print(f"Matrice de poids créée: {w.n} observations, {w.nonzero} liens non-zéros")
print(f"Nombre moyen de voisins: {w.nonzero/w.n:.2f}")

# ===== 3. CALCUL DE L'INDICE DE MORAN GLOBAL =====
# Pour chaque cluster, créer une variable binaire
clusters_uniques = sorted(gdf_clusters[col_cluster].unique())
moran_results = {}

for cluster_id in clusters_uniques:
    # Créer une variable binaire (1 si appartient au cluster, 0 sinon)
    cluster_binary = (gdf_clusters[col_cluster] == cluster_id).astype(int)
    
    # Calculer l'indice de Moran
    moran = Moran(cluster_binary, w)
    
    moran_results[cluster_id] = {
        'I': moran.I,
        'p_value': moran.p_sim,
        'expected_I': moran.EI,
        'variance': moran.VI_sim,
        'z_score': (moran.I - moran.EI) / np.sqrt(moran.VI_sim)
    }
    
    print(f"\nCluster {cluster_id}:")
    print(f"  Indice de Moran I: {moran.I:.4f}")
    print(f"  P-value: {moran.p_sim:.4f}")
    print(f"  Z-score: {moran_results[cluster_id]['z_score']:.4f}")
    
    if moran.p_sim < 0.05:
        if moran.I > 0:
            print(f"  → Autocorrélation spatiale POSITIVE significative")
        else:
            print(f"  → Autocorrélation spatiale NÉGATIVE significative")
    else:
        print(f"  → Pas d'autocorrélation spatiale significative")

# ===== 4. CALCUL DE L'INDICE DE MORAN LOCAL (LISA) =====
# On peut faire l'analyse LISA pour le cluster le plus significatif
cluster_principal = max(moran_results.keys(), 
                       key=lambda x: abs(moran_results[x]['z_score']))

print(f"\nAnalyse LISA pour le cluster {cluster_principal} (plus significatif)")

cluster_binary_principal = (gdf_clusters[col_cluster] == cluster_principal).astype(int)
lisa = Moran_Local(cluster_binary_principal, w)

# Ajouter les résultats LISA au GeoDataFrame
gdf_clusters['lisa_cluster'] = np.nan
gdf_clusters['lisa_p_value'] = np.nan
gdf_clusters['lisa_quadrant'] = np.nan

gdf_clusters.loc[:, 'lisa_cluster'] = lisa.q
gdf_clusters.loc[:, 'lisa_p_value'] = lisa.p_sim

# Classification des quadrants LISA
quadrant_labels = {1: 'HH (High-High)', 2: 'LH (Low-High)', 
                   3: 'LL (Low-Low)', 4: 'HL (High-Low)'}
gdf_clusters['lisa_quadrant'] = gdf_clusters['lisa_cluster'].map(quadrant_labels)

# Points significatifs seulement
gdf_clusters['lisa_significant'] = gdf_clusters['lisa_p_value'] < 0.05

# ===== 5. DIAGRAMME DE MORAN =====
def create_moran_plot(y_values, w_matrix, title="Diagramme de Moran"):
    # Centrer les valeurs
    y_centered = y_values - np.mean(y_values)
    
    # Calculer les moyennes pondérées des voisins
    wy = weights.lag_spatial(w_matrix, y_centered)
    
    # Créer le graphique
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Nuage de points
    ax.scatter(y_centered, wy, alpha=0.6, s=30)
    
    # Régression linéaire
    z = np.polyfit(y_centered, wy, 1)
    p = np.poly1d(z)
    ax.plot(y_centered, p(y_centered), "r--", alpha=0.8, linewidth=2)
    
    # Lignes de référence
    ax.axhline(y=0, color='black', linestyle='-', alpha=0.3)
    ax.axvline(x=0, color='black', linestyle='-', alpha=0.3)
    
    # Annotations des quadrants
    ax.text(0.02, 0.98, 'Q4: LH\n(Low-High)', transform=ax.transAxes, 
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))
    ax.text(0.98, 0.98, 'Q1: HH\n(High-High)', transform=ax.transAxes, 
            verticalalignment='top', horizontalalignment='right', 
            bbox=dict(boxstyle='round', facecolor='lightcoral', alpha=0.5))
    ax.text(0.02, 0.02, 'Q3: LL\n(Low-Low)', transform=ax.transAxes, 
            bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))
    ax.text(0.98, 0.02, 'Q2: HL\n(High-Low)', transform=ax.transAxes, 
            horizontalalignment='right', bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))
    
    ax.set_xlabel('Variable centrée (y)')
    ax.set_ylabel('Retard spatial moyen (Wy)')
    ax.set_title(title)
    ax.grid(True, alpha=0.3)
    
    # Calculer et afficher l'indice de Moran
    moran_i = np.corrcoef(y_centered, wy)[0, 1]
    ax.text(0.02, 0.90, f'Indice de Moran I = {moran_i:.4f}', 
            transform=ax.transAxes, bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    plt.tight_layout()
    return fig

# Créer le diagramme de Moran pour le cluster principal
fig_moran = create_moran_plot(cluster_binary_principal, w, 
                             f"Diagramme de Moran - Cluster {cluster_principal}")
plt.show()

# ===== 6. CARTOGRAPHIE =====
# Créer une figure avec plusieurs sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('Analyse spatiale des clusters en Île-de-France', fontsize=16, fontweight='bold')

# Carte 1: Distribution des clusters
ax1 = axes[0, 0]
gdf_idf_region.boundary.plot(ax=ax1, color='black', linewidth=2)
gdf_idf_departements.boundary.plot(ax=ax1, color='gray', linewidth=1, alpha=0.7)
gdf_clusters.plot(column=col_cluster, categorical=True, legend=True, ax=ax1, 
                 markersize=20, alpha=0.7, cmap='Set1')
ax1.set_title('Distribution des clusters', fontweight='bold')
ax1.set_axis_off()

# Carte 2: Résultats LISA (points chauds/froids)
ax2 = axes[0, 1]
gdf_idf_region.boundary.plot(ax=ax2, color='black', linewidth=2)
gdf_idf_departements.boundary.plot(ax=ax2, color='gray', linewidth=1, alpha=0.7)

# Filtrer les points significatifs
gdf_lisa_sig = gdf_clusters[gdf_clusters['lisa_significant']]
if len(gdf_lisa_sig) > 0:
    gdf_lisa_sig.plot(column='lisa_quadrant', categorical=True, legend=True, ax=ax2,
                     markersize=30, alpha=0.8)
    ax2.set_title(f'Analyse LISA - Points significatifs (p < 0.05)\nCluster {cluster_principal}', 
                 fontweight='bold')
else:
    ax2.text(0.5, 0.5, 'Aucun point significatif\ndans l\'analyse LISA', 
             transform=ax2.transAxes, ha='center', va='center', fontsize=12)
    ax2.set_title(f'Analyse LISA - Cluster {cluster_principal}', fontweight='bold')
ax2.set_axis_off()

# Carte 3: Densité des observations par cluster
ax3 = axes[1, 0]
gdf_idf_region.boundary.plot(ax=ax3, color='black', linewidth=2)
gdf_idf_departements.boundary.plot(ax=ax3, color='gray', linewidth=1, alpha=0.7)
gdf_clusters.plot(column=col_cluster, categorical=True, ax=ax3, markersize=15, alpha=0.6)
ax3.set_title('Vue d\'ensemble avec contours administratifs', fontweight='bold')
ax3.set_axis_off()

# Graphique 4: Résumé statistique
ax4 = axes[1, 1]
ax4.axis('off')

# Tableau des résultats Moran
results_text = "RÉSULTATS INDICE DE MORAN GLOBAL\n" + "="*40 + "\n"
for cluster_id, results in moran_results.items():
    significance = "***" if results['p_value'] < 0.001 else \
                   "**" if results['p_value'] < 0.01 else \
                   "*" if results['p_value'] < 0.05 else ""
    
    results_text += f"Cluster {cluster_id}:\n"
    results_text += f"  I = {results['I']:.4f} {significance}\n"
    results_text += f"  p-value = {results['p_value']:.4f}\n"
    results_text += f"  z-score = {results['z_score']:.4f}\n\n"

results_text += f"Distance de voisinage: {distance_threshold/1000} km\n"
results_text += f"Nombre moyen de voisins: {w.nonzero/w.n:.2f}\n\n"
results_text += "Légende:\n*** p<0.001, ** p<0.01, * p<0.05"

ax4.text(0.1, 0.9, results_text, transform=ax4.transAxes, fontsize=10, 
         verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

plt.tight_layout()
plt.show()

# ===== 7. SAUVEGARDE DES RÉSULTATS =====
# Sauvegarder le GeoDataFrame avec les résultats LISA
output_file = f"clusters_analyse_spatiale_results.shp"
gdf_clusters.to_file(output_file)
print(f"\nRésultats sauvegardés dans: {output_file}")

# Créer un rapport statistique
report = pd.DataFrame(moran_results).T
report.index.name = 'Cluster'
report.to_csv('moran_statistics_report.csv')
print("Rapport statistique sauvegardé dans: moran_statistics_report.csv")
```
